```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Drop "Gender" column if it exists
if "Gender" in df.columns:
    df = df.drop(columns=["Gender"])
```

```python
# Identify categorical columns (excluding Gender)
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
df_encoded = df.copy()  
for col in categorical_cols:
    df_encoded[col] = df_encoded[col].astype('category').cat.codes
```

```python
correlation_matrix = df_encoded.corr()
```

```python
print("Correlation Matrix (Excluding Gender):\n", correlation_matrix)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (Excluding Gender)")
plt.show()
```

```python
# Boxplot to detect outliers
plt.figure(figsize=(10,6))
df.boxplot(rot=45)
plt.title("Boxplot for Outlier Detection")
plt.show()
```

```python
# Encoding Categorical Features
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
from sklearn.preprocessing import LabelEncoder 
# Using Label Encoding (for ordinal variables)
label_encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
# One-Hot Encoding (for nominal variables)
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
# Display transformed dataset
print("\nTransformed Dataset (After Encoding):")
print(df.head())
```
